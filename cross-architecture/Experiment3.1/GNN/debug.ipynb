```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import networkx as nx
import pickle
from tqdm import tqdm
from torch_geometric.utils import from_networkx

# 使用檔案路徑
TRAIN_CSV = "/home/tommy/Projects/cross-architecture/Experiment3.1/dataset/20250509_train.csv"
ROOT_TRAIN_DIR = "/home/tommy/Projects/cross-architecture/Vector/20250509_train"

def quick_check_gpickle(num_samples=10):
    """只載入少量樣本檢查"""
    import pandas as pd
    
    # 讀取CSV檔案
    df = pd.read_csv(TRAIN_CSV)
    file_names = df['file_name'].tolist()[:num_samples]  # 只取前num_samples個
    
    # 創建標籤字典
    labels_dict = {}
    if 'label' in df.columns:
        for idx, row in df.iterrows():
            if idx >= num_samples:
                break
            labels_dict[row['file_name']] = row['label']
    
    print(f"將載入 {num_samples} 個樣本進行檢查")
    
    graphs = []
    loaded_files = []
    
    # 載入gpickle檔案
    for file_name in tqdm(file_names, desc="載入Gpickle檔案"):
        # 嘗試不同的目錄結構
        paths_to_try = [
            # 使用benign/malware子目錄，帶2字元前綴
            *[Path(ROOT_TRAIN_DIR) / sub / file_name[:2] / f"{file_name}.gpickle" 
              for sub in ("benign", "malware")],
            # 使用benign/malware子目錄，不帶前綴
            *[Path(ROOT_TRAIN_DIR) / sub / f"{file_name}.gpickle" 
              for sub in ("benign", "malware")],
            # 直接在根目錄
            Path(ROOT_TRAIN_DIR) / f"{file_name}.gpickle",
        ]
        
        # 嘗試每一個路徑
        for path in paths_to_try:
            if path.exists():
                try:
                    with open(path, "rb") as fp:
                        G = pickle.load(fp)
                    
                    # 成功載入，添加到列表
                    graphs.append(G)
                    loaded_files.append(file_name)
                    
                    # 記錄並打印此圖的信息
                    label = labels_dict.get(file_name, "未知")
                    print(f"\n檔案: {file_name}")
                    print(f"路徑: {path}")
                    print(f"標籤: {label}")
                    print(f"節點數: {G.number_of_nodes()}")
                    print(f"邊數: {G.number_of_edges()}")
                    
                    # 檢查節點屬性
                    if G.nodes():
                        first_node = list(G.nodes())[0]
                        print(f"\n第一個節點 {first_node} 的屬性:")
                        for key, value in G.nodes[first_node].items():
                            if key == 'vector':
                                vector = value
                                print(f"  vector: 長度={len(vector) if hasattr(vector, '__len__') else 'N/A'}, "
                                      f"類型={type(vector).__name__}")
                                if isinstance(vector, np.ndarray) and len(vector) > 0:
                                    print(f"  向量樣本: [{vector[0]:.4f}, {vector[1]:.4f}, ...]")
                                    print(f"  向量統計: 均值={np.mean(vector):.4f}, 標準差={np.std(vector):.4f}")
                            elif key == 'pcode':
                                pcode = value
                                print(f"  pcode: 類型={type(pcode).__name__}")
                                if isinstance(pcode, list) and len(pcode) > 0:
                                    print(f"  pcode樣本: '{pcode[0][:50]}...'")
                                    print(f"  pcode長度: {len(pcode)}")
                            else:
                                print(f"  {key}: {value}")
                    
                    # 轉換為PyG數據對象，檢查兼容性
                    try:
                        # 移除pcode (如果存在)
                        for node in G.nodes():
                            if 'pcode' in G.nodes[node]:
                                del G.nodes[node]['pcode']
                        
                        data = from_networkx(G, group_node_attrs=['vector'])
                        print(f"\n成功轉換為PyG數據對象:")
                        print(f"  節點特徵 shape: {data.x.shape if hasattr(data, 'x') else 'N/A'}")
                        print(f"  邊索引 shape: {data.edge_index.shape if hasattr(data, 'edge_index') else 'N/A'}")
                    except Exception as e:
                        print(f"\n轉換為PyG數據對象失敗: {e}")
                    
                    break  # 找到並成功載入，繼續下一個檔案
                except Exception as e:
                    print(f"[錯誤] 載入Gpickle失敗 {path}: {e}")
        else:
            # 所有路徑都不成功
            print(f"[警告] 找不到檔案: {file_name}.gpickle")
    
    print(f"\n共成功載入 {len(graphs)} 個檔案")
    return graphs, loaded_files

# 執行檢查
graphs, loaded_files = quick_check_gpickle(10)
```
